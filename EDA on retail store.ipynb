```python
# visualizing the data
```

```python
sns.countplot(x=dt['Segment'])
```

```python
dt['Region'].value_counts().plot.pie()
```

```python
print(dt['Sub-Category'].value_counts())
sns.countplot(x=dt['Sub-Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.lineplot(x=dt['Sub-Category'], y=dt['Sales'])
```

```python
sns.pairplot(dt,hue='Category' )
```

```python
sns.heatmap(dt.corr(),annot=True)
```

```python
dt.groupby('Category')[['Sales','Profit']].sum().sort_values(by='Profit', ascending=False).plot.bar(color=['blue','brown'])
plt.ylabel('Sales and Profit')
```

```python
dt.groupby('Region')[['Sales','Profit']].sum().sort_values(by='Profit', ascending=False).plot.bar(color=['blue','brown'])
plt.ylabel('Sales and Profit')
```

```python
sns.lineplot(x='Discount',y='Profit', label='Discount',data=dt)
plt.legend()
plt.show()
```
